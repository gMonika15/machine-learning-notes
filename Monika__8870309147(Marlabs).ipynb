# Getting Dummies

```python
### since we have the categorical columns in dataframe we are creating the dummies columns
df1=pd.get_dummies(df,drop_first=False)
```

```python
df1.head()
```

```python
df1.columns
```

# Finding the top 20 correlation

```python
plt.figure(figsize=[12,10])
corrmat = df1.corr()
k = 20 #number of variables for heatmap
cols = corrmat.nlargest(k, 'RainTomorrow')['RainTomorrow'].index
print(cols)
cm = np.corrcoef(df1[cols].values.T)
sns.set(font_scale=1)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 12}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

# Modelling

```python
## Dropping the high correlated variables
```

```python
x=df1.drop(['RainTomorrow','Humidity3pm','Cloud3pm','Cloud9am','RainToday_Yes','Humidity9am','Rainfall','WindGustSpeed','WindSpeed3pm',
           'MinTemp','WindSpeed9am','Location_Portland','WindDir9am_N','WindDir3pm_WNW','WindGustDir_WNW','Location_Cairns','WindDir9am_NNW',
           'Location_MountGambier','Location_NorfolkIsland','Location_CoffsHarbour'],axis=1)
y=df1[['RainTomorrow']]
x.columns
```

```python
## without dropping the attributes
x1=df1.drop('RainTomorrow',axis=1)
y1=df1.RainTomorrow
```

# Logit summary

```python
# checking the Logit summary to remove the Insignificant variable

from statsmodels.api import add_constant 
import statsmodels.discrete.discrete_model as sm
xa=add_constant(x) #adding a constant as intercept else the line is forced plotted thru origin
log_it=sm.Logit(y,xa)
r=log_it.fit()
r.summary()
```

# Train and Test Split

```python
from sklearn.model_selection import  train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x1,y1,test_size=0.3, random_state=0)
xtrain,xval,ytrain,yval=train_test_split(xtrain,ytrain,test_size=0.3,random_state=0)
```

# Logistic Regression

```python
lor=LogisticRegression()
lor.fit(xtrain,ytrain)
#predict the train data
pred_lor=lor.predict(xtest)#predict has inbuild thershold
print('The model accuracy for logistic Regression validation data is:',round(lor.score(xval,yval)*100,2))
print('The model accuracy for logistic Regression Test data is:',round(lor.score(xtest,ytest)*100,2))
```

# Decision Tree

```python
dtmod=DecisionTreeClassifier()
dtmod.fit(xtrain,ytrain)
pred_dt=dtmod.predict(xtest)
print("The model accuracy for Decision Tree Val data is :",round(dtmod.score(xval,yval)*100,2))
print("The model accuracy for Decision Tree test data is :",round(dtmod.score(xtest,ytest)*100,2))
```

# Random forest

```python
rf=RandomForestClassifier()
rf.fit(xtrain,ytrain)
pred_rf=rf.predict(xtest)
print("\nThe model accuracy for Random Forest Val data is :",round(rf.score(xval,yval)*100,2))
print("The model accuracy for Random Forest test data is :",round(rf.score(xtest,ytest)*100,2))
```
