# scratter plot

```python
sns.lmplot(x='total_bill',y='tip',scatter=True,data=tips,fit_reg=False,height=6)
plt.show()
#fit_reg is true means it gives regression line
```

```python
sns.lmplot(x='total_bill',y='tip',hue='day',data=tips,scatter=True,fit_reg=False,palette='magma',markers=['o','x','v','^'], height=4)

#hue----variable that defines subsets of the data
```

```python
#possible value of palette:------
#accent
#paired
#coolwarm
#rainbow
#spring
#summer
#icefire
```

```python
sns.lmplot(x='total_bill',y='tip',hue='day',col='day',data=tips,scatter=True,fit_reg=False,palette='magma',markers=['o','x','v','^'], height=4)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',col='day',row='sex',data=tips,scatter=True,fit_reg=False,palette='magma', height=4)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,palette='summer')#jitter=False)
plt.show()
```

```python
sns.swarmplot(y='day',x='total_bill',hue='smoker',data=tips,palette='spring',orient='H')#perform the density of the graph
plt.show()
```

```python
sns.countplot(x='day',hue='sex',data=tips,palette='ocean')#count the number of person in differt days
plt.show()
```

```python
sns.countplot(x='size',hue='time',data=tips,palette='spring')
plt.show()
```

```python
sns.countplot(x='smoker',hue='sex',data=tips,palette='summer',saturation=10)
plt.show()
```

```python
#KERNEL DENSITY ESTIMATION KDE
#univariable
sns.distplot(tips['total_bill'])#bell shaped ideal graph
plt.show()
#the entire probabilty is 1
```

```python
#skew hastwo types
#leftskew--------------
#rightskew--------which side has largerslope
```

```python
#bivariable use jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,height=6)
plt.show()
```
