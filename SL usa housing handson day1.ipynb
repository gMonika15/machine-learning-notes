```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('USA_Housing.csv')
```

```python
df1=pd.DataFrame(df)
df1.head()
```

```python
df1.isnull().sum()
```

```python
linreg=LinearRegression()
linreg.fit(pd.DataFrame(df1[['Avg. Area Income','Avg. Area Number of Rooms']]),df1['Price'])
```

# predicted value

```python
linreg.predict(pd.DataFrame(df1[['Avg. Area Income','Avg. Area Number of Rooms']]))
```

```python
df1['predicted values']=linreg.predict(pd.DataFrame(df1[['Avg. Area Income','Avg. Area Number of Rooms']]))
```

```python
df1
```

```python
SSE=sum((df['Price']-df['predicted values'])**2)
print(SSE)
```

```python
n=df1.shape[0]
n
```

```python
RMSE=np.sqrt(SSE/n)
print(RMSE)
```

```python
MAPE=(sum(abs((df1['Price']-df1['predicted values'])/df1['Price'])))/n
print(MAPE)
```

```python
TSS=sum((df1['Price']-np.mean(df1['Price']))**2)
print(TSS)
```

```python
R2=1-(SSE/TSS)# R value will be between 0to 1
print(R2)
```
